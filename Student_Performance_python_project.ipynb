```python
df.drop('Maths_Passing_Marks', inplace = True, axis = 1)
df.drop('Reading_Passing_Marks', inplace = True, axis = 1)
df.drop('Writing_Passing_Marks', inplace = True, axis = 1)
```

```python
sex = pd.get_dummies(df['gender'], drop_first = True)
group = pd.get_dummies(df['groups'], drop_first = True)
education = pd.get_dummies(df['educaion level'], drop_first = True)
course = pd.get_dummies(df['test preparation course'], drop_first = True)
```

```python
df = pd.concat([df, sex, group, education, course], axis = 1)
df.head()
```

```python
df[df.select_dtypes(['bool']).columns] = df.select_dtypes(['bool']).astype(int)
print(df)
```

```python
df.drop(['gender', 'groups', 'educaion level', 'test preparation course' ], axis = 1, inplace = True)
print('train_data shape',df.shape)
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.drop(columns = ['male'], axis=1)
y = df['male']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .20, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
lreg = LogisticRegression()
lreg.fit(x_train, y_train)
```

```python
y_pred_lreg = lreg.predict(x_test)
print(y_pred_lreg)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print('Classification Model')
print('--'*20)
logreg_accuracy = round(accuracy_score(y_test, y_pred_lreg) * 100,2)
print('Accuracy is : ', logreg_accuracy,'%')
```
